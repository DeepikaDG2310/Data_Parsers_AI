```python
def analyse_doc(elements):
    eletype_count={}

    for i , e in enumerate(elements):
        eletype = type(e).__name__
        eletype_count[eletype] = eletype_count.get(eletype, 0)+1

    for e,c in eletype_count.items():
        print(f"{e} : {c}")
    
    return eletype_count
```
